```python
import yfinance as yf
import pandas as pd

# Create a Ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Extract historical stock data (let's use a period of 'max' to get all available data)
tesla_data = tesla.history(period="max")

# Reset the index to make 'Date' a column
tesla_data.reset_index(inplace=True)

# Save the dataframe (optional, if you need to save it to a file)
tesla_data.to_csv("tesla_data.csv", index=False)

# Display the first five rows
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for Tesla revenue data (example, you may need to find the correct one)
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Send a GET request to the webpage
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

# Parse the webpage content
soup = BeautifulSoup(response.text, "html.parser")

# Find the table containing revenue data (youâ€™ll need to inspect the website for the correct selectors)
tables = soup.find_all("table")
tesla_revenue_table = None
for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        tesla_revenue_table = table
        break

# Extract data into a DataFrame
data = []
rows = tesla_revenue_table.find_all("tr")[1:]  # Skip header row
for row in rows:
    cols = row.find_all("td")
    date = cols[0].text.strip()
    revenue = cols[1].text.strip().replace("$", "").replace(",", "")  # Clean revenue data
    if revenue:  # Only add rows with revenue data
        data.append([date, revenue])

# Create DataFrame
tesla_revenue = pd.DataFrame(data, columns=["Date", "Revenue"])

# Convert Revenue to numeric (if needed)
tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"], errors="coerce")

# Display the last five rows
print(tesla_revenue.tail())
```

```python
import plotly.graph_objects as go

def make_graph(stock_data, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=stock_data["Date"], y=stock_data["Close"], mode="lines", name="Close Price"))
    fig.update_layout(
        title=title,
        xaxis_title="Date",
        yaxis_title="Price (USD)",
        template="plotly_dark"
    )
    fig.show()

# Plot Tesla stock data (limit to recent data for clarity, e.g., last year)
tesla_data_recent = tesla_data[tesla_data["Date"] >= "2023-01-01"]
make_graph(tesla_data_recent, "Tesla Stock Price (2023-2024)")
```

```python
# Plot GameStop stock data
gme_data_recent = gme_data[gme_data["Date"] >= "2023-01-01"]
make_graph(gme_data_recent, "GameStop Stock Price (2023-2024)")
```
